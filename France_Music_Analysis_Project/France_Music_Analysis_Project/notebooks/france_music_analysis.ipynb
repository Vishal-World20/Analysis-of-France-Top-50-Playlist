# France Top 50 Playlist Analysis

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/Atlantic_France.csv')

df.head()
```

```
df['date'] = pd.to_datetime(df['date'])
df['duration_min'] = df['duration_ms'] / 60000
df['album_type'] = df['album_type'].str.lower().str.strip()

df.info()
```

## Explicit Content Analysis

```
explicit_share = df['is_explicit'].value_counts(normalize=True) * 100

explicit_share.plot(kind='pie', autopct='%1.1f%%')
plt.title('Explicit vs Clean Content')
plt.ylabel('')
plt.show()
```

## Album Type Analysis

```
sns.countplot(x='album_type', data=df)
plt.title('Album Type Distribution')
plt.show()
```

## Duration Analysis

```
sns.histplot(df['duration_min'], bins=30, kde=True)
plt.title('Song Duration Distribution')
plt.show()
```

## Correlation Analysis

```
numeric_cols = ['position', 'popularity', 'duration_min', 'total_tracks']

corr = df[numeric_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
